```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
req=requests.get('https://coinmarketcap.com/currencies/bitcoin/historical-data/?start=20130428&end=20180124')
```

```python
req
```

```python
html=req.text
```

```python
html
```

```python
soup=BeautifulSoup(html,'html.parser')
```

```python
soup
```

```python
columns=soup.select('div.table-responsive > table > thead > tr > th')
```

```python
columnlist=[]
for column in columns:
    columnlist.append(column.text)
df=pd.DataFrame(columns=columnlist)
df
```

```python
contents=soup.select('div.table-responsive > table > tbody > tr ')
#contents[0].text.strip('\n').split('\n')
dfcontent=[]
alldfcontents=[]
contents[0].find_all("td")
```

```python
for content in contents:
    tds=content.find_all("td")
    for td in tds:
        dfcontent.append(td.text)
    alldfcontents.append(dfcontent)
    dfcontent=[]
        
        
```

```python
alldfcontents
```

```python
df=pd.DataFrame(columns=columnlist, data=alldfcontents)
```

```python
df
```

```python
df.dtypes
```

```python
df['Close']=pd.to_numeric(df.Close,errors='coerce')
```

```python
df.dtypes
```

```python
newMarketCap=[]
for number in df['Market Cap']:
    lst=number.split(',')
    total=''
    for i in lst:
        total+=str(i)
    newMarketCap.append(int(total))

    
```

```python
df['Market Cap']=pd.DataFrame(newMarketCap)
```
